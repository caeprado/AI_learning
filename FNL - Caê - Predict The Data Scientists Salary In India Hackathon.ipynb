#Limpeza de Colunas

###Melhornado código base

```python
#começar a corrigir as informações das colunas

#analisando as informações, o que podemos considerar importante para a criação da máquina
#são as colunas key_skills, job_desig, job_type, salary e location
#para isso é necessário fazer uma limpeza nas linhas para que a informação fiquei padronizada
#pesquisando encontrei uma forma criando uma def e transfomando a base em um df


data_train = data_train.dropna(subset=['key_skills']) #eliminando uma linha vazia da coluna
#cria com a def uma função lambda para limpar as linhas da base
def clean_skills(skl):
    skills = str(skl).lower()
    skills = re.sub(r'\W', ' ', skills)  # Substitui não palavras por espaços
    skills = re.sub(r'\s+', ' ', skills)  # Remove espaços em excesso
    return skills.strip()  # Remove espaços no início e no final

# Cria os dataframes
df_train = data_train[['key_skills', 'job_desig', 'job_description', 'location', 'job_type', 'experience', 'salary']]
df_test = data_test[['key_skills', 'job_desig', 'job_description', 'job_type', 'experience', 'location']]

# Aplica a limpeza nas colunas 'key_skills' dos dataframes
df_train['skills_cleaned'] = df_train['key_skills'].apply(clean_skills)
df_test['skills_cleaned'] = df_test['key_skills'].apply(clean_skills)
```

```python
df_train.head()
```

```python
#Limpando a coluna job_design
def clean_job_desig(desig):
    job_desig = desig.lower()
    job_desig = re.sub(r'[^a-z]', ' ', job_desig)
    job_desig = re.sub(r'\s+', ' ', job_desig)
    return job_desig

df_train['desig_cleaned'] = df_train['job_desig'].apply(clean_job_desig)
df_test['desig_cleaned'] = df_test['job_desig'].apply(clean_job_desig)

df_train.head()
```

```python
#Limpando a coluna job_design
#na coluna job_description tem muitos valores nulos, colocar os valores vazios como missing
data_train['job_description'].fillna('missing', inplace=True)
data_test['job_description'].fillna('missing', inplace=True)

def clean_job_desc(job):
    job_desc = str(job).lower()
    job_desc = re.sub(r'[^a-z]', ' ', job_desc)
    job_desc = re.sub(r'\s+', ' ', job_desc)
    return job_desc

df_train['job_desc_cleaned'] = df_train['job_description'].apply(clean_job_desc)
df_test['job_desc_cleaned'] = df_test['job_description'].apply(clean_job_desc)

df_train.head()
```

```python
#seguindo a mesma base do anterior

def clean_location(loc):
    location = loc.lower()
    location = re.sub(r'[^a-z]', ' ', location)
    location = re.sub(r'\s+', ' ', location)
    return location

df_train['location_cleaned'] = df_train['location'].apply(clean_location)
df_test['location_cleaned'] = df_test['location'].apply(clean_location)

df_train.head()
```

```python
#o job_values tem problema de NaN e diferentes formas de escrever analytics, o que foi padronizado abaixo
job_type_values = ['Analytics', 'Analytic', 'ANALYTICS', 'analytic']
replacement_value = 'analytics'

data_train['job_type'] = data_train['job_type'].fillna('missingjobtype').replace(job_type_values, replacement_value)
data_test['job_type'] = data_test['job_type'].fillna('missingjobtype').replace(job_type_values, replacement_value)

df_train.head()
```

```python
#separando a coluna de mínimo e máximo para facilitar a leitura do algorítimo e criando a merde depois

def min_exp(exp):
    val = re.sub(r'\-', ' ', exp)
    val = val.split(' ')
    val = int(val[0])
    return val

def max_exp(exp):
    val = re.sub(r'\-', ' ', exp)
    val = val.split(' ')
    if len(val) > 1:
        val = int(val[1])
    else:
        val = None
    return val

df_train['min_exp'] = df_train['experience'].apply(min_exp)
df_train['max_exp'] = df_train['experience'].apply(max_exp)
df_test['min_exp'] = df_test['experience'].apply(min_exp)
df_test['max_exp'] = df_test['experience'].apply(max_exp)

df_train['merged'] = df_train['desig_cleaned'] + ' ' + df_train['job_desc_cleaned'] + ' ' + df_train['skills_cleaned'] + ' ' + df_train['job_type']
df_test['merged'] = df_test['desig_cleaned'] + ' ' + df_test['job_desc_cleaned'] + ' ' + df_test['skills_cleaned'] + ' ' + df_test['job_type']

df_train.head()
```

#Criando a máquina virtual

```python
#Classificar cada valor único da base para facilitar a visualização

le = LabelEncoder()
df_train['salary'] = le.fit_transform(df_train['salary'])
le.classes_
```

```python
#Determinando quais são as variáveis determinantes e preenchendo as linhas vazias para não ficarem sem valores
#nesse caso o valor nulo - "" é diferente de sem valor NaN

df_train['merged'].fillna('', inplace=True)
df_train['location_cleaned'].fillna('', inplace=True)
df_test['merged'].fillna('', inplace=True)
df_test['location_cleaned'].fillna('', inplace=True)

train_features, validation_features, train_labels, validation_labels = train_test_split(
    df_train[['merged', 'location_cleaned', 'min_exp', 'max_exp']],
    df_train['salary'], test_size=0.20,
    stratify=df_train['salary'], random_state=75)

print('No. of sample texts in train_features:', len(train_features))
print('No. of sample texts in validation_features:', len(validation_features))
print('No. of sample labels in train_labels:', len(train_labels))
print('No. of sample labels in validation_labels:', len(validation_labels))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from scipy.sparse import hstack, csr_matrix
from sklearn.preprocessing import StandardScaler
from scipy import sparse

# Unificando as colunas 'merged' e 'location_cleaned'
train_features_merged = train_features['merged'].fillna('')
train_features_loc = train_features['location_cleaned'].fillna('')
validation_features_merged = validation_features['merged'].fillna('')
validation_features_loc = validation_features['location_cleaned'].fillna('')
combined_train = train_features_merged + ' ' + train_features_loc
combined_validation = validation_features_merged + ' ' + validation_features_loc

# Aplicando TfidfVectorizer
tf1 = TfidfVectorizer(min_df=3, token_pattern=r'\w{3,}', ngram_range=(1,3), max_df=0.9)
tf2 = TfidfVectorizer(min_df=2, token_pattern=r'\w{3,}')

X_train_merged = tf1.fit_transform(combined_train)
X_train_loc = tf2.fit_transform(train_features_loc)

X_validation_merged = tf1.transform(combined_validation)
X_validation_loc = tf2.transform(validation_features_loc)

# Padronização dos valores numéricos
sc1 = StandardScaler()
X_train_MinExp = sc1.fit_transform(np.array(train_features['min_exp']).reshape(-1, 1))
X_validation_MinExp = sc1.transform(np.array(validation_features['min_exp']).reshape(-1, 1))

sc2 = StandardScaler()
X_train_MaxExp = sc2.fit_transform(np.array(train_features['max_exp']).reshape(-1, 1))
X_validation_MaxExp = sc2.transform(np.array(validation_features['max_exp']).reshape(-1, 1))

# Convertendo para matrizes esparsas
X_train_exp = csr_matrix(X_train_MinExp)
X_validation_exp = csr_matrix(X_validation_MinExp)
X_train_max_exp = csr_matrix(X_train_MaxExp)
X_validation_max_exp = csr_matrix(X_validation_MaxExp)

# Combinação das matrizes esparsas
merged_train = hstack((X_train_merged, X_train_loc, X_train_exp, X_train_max_exp))
merged_validation = hstack((X_validation_merged, X_validation_loc, X_validation_exp, X_validation_max_exp))

print("Shape of merged_train:", merged_train.shape)
print("Shape of merged_validation:", merged_validation.shape)

```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import lightgbm as lgb

#definindo o dataset

train_data = lgb.Dataset(merged_train, label=train_labels)
test_data = lgb.Dataset(merged_validation, label=validation_labels)

# Definindo os parâmetros do modelo
params = {
    'objective': 'multiclass',
    'num_class': 6,
    'metric': 'multi_logloss',
    'num_leaves': 31,
    'learning_rate': 0.05,
    'feature_fraction': 0.9
}

# Treinando o modelo

lgbm = lgb.train(params=param, train_set=train_data, num_boost_round=100, valid_sets=[test_data])

y_pred_class = lgbm.predict(merged_validation)

```

```python


predictions = []
for x in y_pred_class:
    predictions.append(np.argmax(x))

accuracy = accuracy_score(validation_labels, predictions)
print('Accuracy:', accuracy)
```
