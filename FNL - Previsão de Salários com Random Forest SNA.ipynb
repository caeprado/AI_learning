```python
import pandas as pd
from pandas_profiling import ProfileReport
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

#Importação dos dados

```python
#importação dos dados diretamente do github

data = pd.read_csv('https://raw.githubusercontent.com/llSourcell/Best-Programming-Languages-for-Machine-Learning/master/adults.txt', sep=",")
data
```

### Criação de Report com o Pandas Profiling

```python
#criação do relatório uando profiling para visualização melhor dos dados

profile = ProfileReport(data, title='Relatório Base de Dados', html={'style}':{'full_width': True}})

profile.to_notebook_iframe()

profile.to_file(output_file="Relatório Bse de Dados.html")
```

#Pré-Processamento dos daods

```python
#para melhor leitura do algorítimo alterar as variáveis string para numéricas

for variaveis in ['sex', 'race', 'occupation', 'education', 'workclass', 'relationship' ]:
    data[variaveis] = LabelEncoder().fit_transform(data[variaveis])

# olhar data para verificar as alterações
#data
```

```python
#selecionando variáveis a serem utilizadas na previsão e incluindo dentro do X (x dentro da função
# preditiva são as variáveis determinantes e o y a variável dependente)
X = data[['sex', 'race', 'hours_per_week', 'occupation', 'education', 'workclass', 'relationship', 'capital_gain' ]]

#definindo a variável predita (dependente), Y

Y = data['salary'].values.tolist()
```

```python
#dividindo entre dados de treino e teste (30% teste)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
```

#Criando a máquina preditiva

```python
#criando a máquina preditiva - instânciando o algoritimo de ML (chamando o algorítiomo a ser usado)
#clf_RF - classificador random forest
clf_RF = RandomForestClassifier(n_estimators=1000) #escolhendo a quantidade de árvores de decisão
```

```python
#treinando o algorítimo

clf_RF = clf_RF.fit(X_train, Y_train)
```

###Avaliando a Acurácia do Modelo de Machine Learning

```python
#Validando a Máquina Preditiva - Verificando o nível de acurácia

accuracy = clf_RF.score(X_test, Y_test)
print('Accuracy:' + str(accuracy))

#alterar algumas das variáveis preditivas para determinar em qua ponto a acurácia é maior
#baseline acurácia 0.8151294912478247
#acurácia com ganho de capital incluso 0.8440986794963661
```

```python
#Determinando a importância das variáveis

importance = clf_RF.feature_importances_
print(importance)
```

#Fazendo Novas Predições

```python
#Utilizando a Maquina Preditiva - Fazendo previsões com novos dados
#Não temos novos dados, feito usando os daods da base existente
Novas_Previsoes = clf_RF.predict(X_test)
print(Novas_Previsoes)
```
